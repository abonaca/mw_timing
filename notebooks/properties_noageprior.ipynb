## Age distribution

```python
age = 10**t['logAge']*1e-9
age_lerr = age - 10**(t['logAge']-t['logAge_lerr'])*1e-9
age_uerr = 10**(t['logAge']+t['logAge_uerr'])*1e-9 - age
age_err = 0.5 * (age_lerr + age_uerr)
age_err_cons = np.maximum(age_lerr, age_uerr)
age_err_min = 1
age_err_cons[age_err_cons<age_err_min] = age_err_min
age_invar = (age_err_cons)**-2
```

```python
ind_lowe = t['eccen_pot1']<0.3

npop = np.sum(ind_safeage & ind_trap & ind_lowe)
obs_pop = age[ind_safeage & ind_trap & ind_lowe]
err_pop = age_err[ind_safeage & ind_trap & ind_lowe]

np.random.seed(67294)

abins = np.linspace(4,14,20)
abcen = 0.5 * (abins[1:] + abins[:-1])

nboot = 1000
nbin = np.size(abins) - 1
bcen = 0.5 * (abins[1:] + abins[:-1])
h_age_pop = np.empty((nboot, nbin))



for i in range(nboot):
    # sample accreted halo from a flat distribution
    model_pop = np.random.rand(npop)*4 + 6.
    log_model_pop = np.log10(model_pop)
    log_model_pop_conv = (np.random.randn(npop)*
                          np.random.permutation(t['logAge_lerr'][ind_safeage & ind_trap & ind_lowe])
                          + np.random.permutation(log_model_pop))
    model_pop_conv = 10**log_model_pop_conv
    h, be = np.histogram(model_pop_conv, bins=abins, density=True)
    h_age_pop[i] = h

model_pop_dn, model_pop_up = np.percentile(h_age_pop, [16,84], axis=0)


plt.figure(figsize=(11,7))

# plt.hist(obs_ins, bins=abins, color=orange, histtype='step', lw=4, density=True, label='In-situ halo')
# # plt.fill_between(bcen, model_ins1_dn, model_ins1_up, color='darkorange', alpha=0.3, label='In-situ model (7.3 Gyr)')
# plt.fill_between(bcen, model_ins_dn, model_ins_up, color=orange, alpha=0.3, label='In-situ model (8.3 Gyr)')
# # plt.fill_between(bcen, model_ins2_dn, model_ins2_up, color='red', alpha=0.3, label='In-situ model (9.3 Gyr)')

plt.hist(obs_pop, bins=abins, color='b', histtype='step', lw=4, density=True, label='Observed')
plt.fill_between(bcen, model_pop_dn, model_pop_up, color='b', alpha=0.3, label='const SFR $6-10$ Gyr')

# plt.hist(obs_low, bins=abins, color=pink, histtype='step', lw=4, density=True, label='Low $\\alpha$ disk')
# plt.fill_between(bcen, model_low_dn, model_low_up, color=pink, alpha=0.3, label='Low $\\alpha$ model')

plt.legend(loc=1, frameon=False)
plt.xlabel('Age [Gyr]')
plt.ylabel('Density [Gyr$^{-1}$]')

plt.tight_layout()
plt.savefig('../plots/low_alpha_const_sfh.png')
```
